```python
#this for loop will make a separate dataframe for each stock getting data from yahoo finance
for df in stock_list:
    symbol = df
    data_source = 'yahoo'
    start_date = '2020-02-01'
    end_date = '2020-05-01'
    #now storing dataframe of each stock in the stock dictionary created earlier
    stock[df] = data.DataReader(symbol, data_source, start_date, end_date)
    stock[df] = stock[df][['Open', 'High', 'Low', 'Close', 'Adj Close', 'Volume']] #changing positions of columns
    stock[df].name = df
```

```python
#making a function that calculates adjusted closing price between 20 Feb and 23 March and percent decline between them
def stk_adj_c(s):
    tick = s.name
    adj_c_20 = s.loc['2020-02-20']['Adj Close']
    adj_c_23 = s.loc['2020-03-23']['Adj Close']
    adj_c_d = ((adj_c_20-adj_c_23)/adj_c_20)*100
    return tick, adj_c_20, adj_c_23, adj_c_d
```

```python
adj_c_rlc = stk_adj_c(stock['RELIANCE.NS'])
adj_c_rlc
```

```python
#making a fuction that will make a scatterplot of adjusted closing price of stocks that are passed
def plot_s(*args):
    fig = go.Figure() #creating a function of plotly offline
    for stock in args: #it will perform following for each stock in the passed argument
        fig.add_trace(go.Scatter(x=stock.index, y=stock['Adj Close'], mode='lines+markers', name=stock.name, showlegend=True))
    fig.show()    
```

```python
#making a function that will create a dataframe with the adjusted closing stock price of the stocks on 25th Feb and 24th March and the percentage decline that took place in betwwen them for the stocks that are passed as arguments
def df_s(*args):
    temp_adj_list = []
    for stock in args:
        adj_c = stk_adj_c(stock) #getting the adjusted closing price that stock
        temp_adj_list.append(adj_c) #adding the adjusted closing price for that stock in list created
    temp_df = pd.DataFrame(temp_adj_list, columns=['Ticker', 'Adj Close on 25th Feb', 'Adj Close on 24th March', 'Percent Decline'])
    temp_df_comp = temp_df.merge(comp_df, left_on='Ticker', right_on='ticker', how='left')
    return temp_df_comp[['ticker','name','Adj Close on 25th Feb', 'Adj Close on 24th March', 'Percent Decline']]
```
